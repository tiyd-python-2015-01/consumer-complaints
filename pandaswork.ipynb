I got census data with polulation numbers

```python
myfile = "census.csv"
census = pd.read_csv(myfile, dtype={'Population': np.float64} )
census.columns=['State','State_name','Population']
census.set_index(['State'])
```

Getting compalints by state.  Turning a series into a data frame


```python
df
dfstate = df['State'].value_counts()

dfframe = dfstate.to_frame()
dfframe.index

census.set_index = ['State']
census.index

dfframe.columns=["Complaints"]
```

Merging the 2 data sets

```python
all_data=pd.merge(dfframe, census, how='left', left_index=True, right_on=['State'])
all_data.columns= ['Complaints','State','State_name','Population']
all_data['Population'] = all_data['Population'].str.replace(r'[$,]', '').astype('float')
all_data
```

Getting per capita complaints

```python
all_data['per_capita']= all_data['Complaints']/all_data['Population']
```

```python
all_data[['State_name','per_capita']].sort(['per_capita'],ascending=False)
```

Calculating Outliers

```python

outlier_zip=df['ZIP code'].value_counts()
outlierdf = outlier_zip.to_frame()
outlierdf.columns=['Complaints']



newdf = outlierdf.copy()
#newdf.set_index(['ZIP code'])

newdf['x-Mean'] = abs(newdf['Complaints'] - newdf['Complaints'].mean())
newdf['1.96*std'] = 1.96*newdf['Complaints'].std()  
newdf['Outlier'] = abs(newdf['Complaints'] - newdf['Complaints'].mean()) > 1.96*newdf['Complaints'].std()
```
