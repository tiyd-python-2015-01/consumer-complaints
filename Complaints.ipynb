```python
complaints["dates"] = pd.to_datetime(complaints.pop('Date received'), format="%m/%d/%Y")
#The above creates a new column named "dates" by converting the 'Date received' column
#into the proper date time format and then popping it out to be re-assigned
```

```python
weekdays = complaints.dates.dt.weekday
weekdays
```

```python
weekdays_counts = weekdays.value_counts()
day_sort = weekdays_counts.sort_index()
day_sort

#Mon=0, Tues =1, Weds=2, Thurs=3, Fri = 4, Sat=6, Sun=6
```

```python
day_sort.index = ['Mon', 'Tues', 'Weds', 'Thurs', 'Fri', 'Sat', 'Sun']
day_sort
```

```python
dec = cal.monthrange(2014, 12)
dec
#First day of 12/2014 was 0(Monday) and there are 31 days in the month
```

```python
dec1 = cal.monthcalendar(2014, 12)
dec1
#Returns a matrix representing a month’s calendar. Each row represents a week; days outside of the month a represented by zeros.
#Each week begins with Monday unless set by setfirstweekday()
```

```python
freq = [5,5,5,4,4,4,4]
average = day_sort/freq
average
```

```python
average.plot(kind='bar', fontsize=20, figsize=(10, 10))
```
