# Number of Complaints by Day of Week

```python
complaints.index = pd.to_datetime(complaints.pop('Date received'), format="%m/%d/%Y")
```

```python
complaints_by_day = complaints.index.day
complaints_by_day
```

```python
complaints['DOW'] = complaints_by_day
day_counts = complaints.groupby('DOW').size()
day_counts.plot(kind='bar')
plt.show()
```

```python
day_counts.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday",
                    "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday", 
                    "Monday", "Tuesday", "Wednesday"]

day_counts.plot(kind='bar')
plt.show()
```

# Average # of Complaints by Day of the Week

```python
monday = day_counts.Monday.sum()
tuesday = day_counts.Tuesday.sum()
wednesday = day_counts.Wednesday.sum()
thursday = day_counts.Thursday.sum()
friday = day_counts.Friday.sum()
saturday = day_counts.Saturday.sum()
sunday = day_counts.Sunday.sum()
print("Total complaints on Mondays:",monday)
print("Total complaints on Tuesdays:",tuesday)
print("Total complaints on Wednesdays:",wednesday)
print("Total complaints on Thursdays:",thursday)
print("Total complaints on Fridays:",friday)
print("Total complaints on Saturdays:",saturday)
print("Total complaints on Sundays:",sunday)
```

```python
avg_monday_complaints = monday/len(day_counts.Monday)
avg_tuesday_complaints = tuesday/len(day_counts.Tuesday)
avg_wednesday_complaints = wednesday/len(day_counts.Wednesday)
avg_thursday_complaints = thursday/len(day_counts.Thursday)
avg_friday_complaints = friday/len(day_counts.Friday)
avg_saturday_complaints = saturday/len(day_counts.Saturday)
avg_sunday_complaints = sunday/len(day_counts.Sunday)

print("Average # of complaints on Monday:",avg_monday_complaints)
print("Average # of complaints on Tuesday:",avg_tuesday_complaints)
print("Average # of complaints on Wednesday:",avg_wednesday_complaints)
print("Average # of complaints on Thursday:",avg_thursday_complaints)
print("Average # of complaints on Friday:",avg_friday_complaints)
print("Average # of complaints on Saturday:",avg_saturday_complaints)
print("Average # of complaints on Sunday:",avg_sunday_complaints)
```

```python
list_of_complaint_avg = [avg_monday_complaints, avg_tuesday_complaints, avg_wednesday_complaints, avg_thursday_complaints, avg_friday_complaints,
                         avg_saturday_complaints, avg_sunday_complaints]
sizes = list_of_complaint_avg
labels = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
explode = (0, 0.1, 0, 0, 0, 0, 0)
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'blue', 'red', 'orange']
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("% of Complaints by day\n\n", fontsize = 20)
plt.axis('equal')
plt.show()
```
