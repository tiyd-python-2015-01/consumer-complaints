### Weight the Days of Week

```python
complaints['days'] = pd.to_datetime(complaints.pop("Date received"), format='%m/%d/%Y')
```

```python
num_of_complaints = max(complaints.count())
timespan = len(complaints['days'].unique())
dates = complaints['days'].unique()
days = []
for day in dates:
    dt64 = np.datetime64(day)
    ts = (dt64 - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
    days.append(datetime.utcfromtimestamp(ts).weekday())
count_length = len(days)
days_count = Counter(days)
day_ratio = []
print("Monday - Friday Count")
print(count_length)
print(days_count)
for n in range(7):
    day_ratio.append(31/(days_count[n]))
#day_ratio = list(days_count.values())
print(day_ratio)
```

### Number of Complaints per Day of Week

```python
complaints['Days'] = complaints.days.dt.dayofweek
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday',]
day_map = lambda x: days[x]
complaints_by_day = complaints['Days'].map(day_map)
complaints_by_day = complaints_by_day.value_counts()
print(complaints_by_day)
```

### Mean Number of Complaints per Day of Week

```python
complaints_by_day = complaints['Days'].value_counts()/timespan
complaints_by_day
```

```python
complaints_by_day *= day_ratio 
complaints_by_day
```
