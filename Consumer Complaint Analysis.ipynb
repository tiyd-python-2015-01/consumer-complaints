# Averaging Complaints Per Day of Week

```python
complaints['datetime'] = pd.to_datetime(complaints['Date received'])
```

```python
complaints['weekday'] = complaints['datetime'].apply(lambda x: x.weekday())
```

```python
complaints_by_weekday = complaints['weekday'].value_counts()
```

```python
complaints_sorted_days = complaints_by_weekday.sort_index()
```

```python
complaints_sorted_days.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
day_weights = [5,5,5,4,4,4,4]
```

```python
mean_complaints_per_day = complaints_sorted_days / day_weights
```

```python
plt.figure(figsize=(9,9))
plt.title('Average Complaints Per Day')
mean_complaints_per_day.plot(kind='bar', color='#d62e2e')



```
