```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
```

```python
%matplotlib inline
```

```python
complaints = pd.read_csv("complaints_dec_2014.csv")
```

```python
by_company = complaints["Company"].value_counts()[:10]
by_company.plot(kind = "bar")
#sns.factorplot("Company", data = by_company, palette ="muted")
```

```python
by_product = complaints["Product"].value_counts()[:10]
by_product.plot(kind = "bar")
#sns.factorplot("Company", data = by_company, palette ="muted")
```

```python
by_response = complaints["Company response"].value_counts()[:10]
by_response.plot(kind = "bar")
```

```python
complaints["dates"]  = pd.to_datetime(complaints.pop('Date received'), format="%m/%d/%Y")
```

```python
top_days = complaints.dates.dt.weekday.value_counts()
```

```python
top_days = top_days.sort_index(ascending=True)
```

```python
top_days.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
top_days.plot(kind = "bar")
```

```python
import calendar
my_list = []
for n in range(1,32):
    my_list.append(calendar.weekday(2014, 12, n))
```

```python
freqs = np.unique(my_list, return_counts = True)[1]
```

```python
the_averages = top_days/freqs
```

```python
the_averages.plot(kind = "bar")
```

It seems on average Tuesdays have the highest complaints. This may be due to the fact that billing mistakes etc happen on the first day of the week and are noticed on the second day.
